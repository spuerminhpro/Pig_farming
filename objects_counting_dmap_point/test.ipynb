```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import random

# Update the path to point to your HDF5 file (e.g., train.h5)
h5_path = '/mnt/data/PythonProject/Pig_counting/Pig_farming/objects_counting_dmap_point/data/train.h5'

# Open the HDF5 file
with h5py.File(h5_path, 'r') as hf:
    images = hf['images'][:]
    labels = hf['labels'][:]

# Select a random sample
idx = random.randint(0, images.shape[0] - 1)
img = images[idx]
label = labels[idx]

# Compute estimated count from the density map
object_count = np.sum(label)  # Sum of the density map values

# Print shape info and count
print(f"Selected Image shape: {img.shape}")    # Expecting (3, H, W)
print(f"Selected Label shape: {label.shape}")  # Expecting (1, H, W)
print(f"Estimated Object Count: {object_count:.2f}")  # Approximate count

# Convert image from channel-first (3, H, W) to channel-last (H, W, 3)
img_display = np.transpose(img, (1, 2, 0))
# If the label has one channel, squeeze it to 2D
label_display = label.squeeze()

# Display the image and its density map
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_display)
plt.title("Random Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(label_display, cmap='jet')
plt.colorbar(label="Density")
plt.title(f"Density Map (Count: {object_count:.2f})")
plt.axis('off')

plt.show()
```
